```python
df = pd.read_csv('garments_worker_productivity.csv')
df.drop("wip", axis=1, inplace=True)
df.drop("smv", axis=1, inplace=True)
df.drop("quarter", axis=1, inplace=True)
df.drop("date", axis=1, inplace=True)

df["day"] = df.day.astype("category")
df["day"] = df["day"].cat.rename_categories({"Monday": 1, "Tuesday": 2, "Wednesday": 3, "Thursday": 4, "Saturday": 6, "Sunday": 7})
df["day"] = df.day.astype("int64")

# Locate and apply attribute category correction
df["department"] = df.department.astype("object")
df["department"].mask(df["department"] == "finishing ", "finishing", inplace=True)
df["department"].mask(df["department"] == "sweing", "sewing", inplace=True)
df["department"] = df.department.astype("category")
df["department"] = df["department"].cat.rename_categories({"finishing": 1, "sewing": 2})
df["department"] = df.department.astype("int64")

# Reassign incentive values to (No) incentive and (Yes) incentive
df["incentive"] = df.incentive.astype("object")
df["incentive"].mask(df["incentive"] > 0, "yes", inplace=True)
df["incentive"].mask(df["incentive"] == 0, "no", inplace=True)
df["incentive"] = df["incentive"].astype("category")
df["incentive"] = df["incentive"].cat.rename_categories({"yes": 1, "no": 0})
df["incentive"] = df.incentive.astype("int64")

# Reassign Number of style changes to (No) style change and (Yes) style change
df["no_of_style_change"] = df.no_of_style_change.astype("object")
df["no_of_style_change"].mask(df["no_of_style_change"] > 0, "yes", inplace=True)
df["no_of_style_change"].mask(df["no_of_style_change"] == 0, "no", inplace=True)
df["no_of_style_change"] = df["no_of_style_change"].astype("category")
df["no_of_style_change"] = df["no_of_style_change"].cat.rename_categories({"yes": 1, "no": 0})
df["no_of_style_change"] = df.no_of_style_change.astype("int64")

df['productivity_difference'] = df['targeted_productivity'] - df['actual_productivity']

print(df.dtypes)
df.head()
```

```python
# Import necessary libraries
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Descriptive Statistics
summary = df1.describe()

# Function to plot subplots
def plot_subplots(plot_type, df):
    columns = df.select_dtypes(include=['int64', 'float64']).columns
    n_columns = 3  # number of plots per row
    n_rows = 4  # number of plots per column

    # Total number of subplots
    total_plots = n_columns * n_rows

    # Number of pages needed
    n_pages = -(-len(columns) // total_plots)  # Ceiling division

    for page in range(n_pages):
        # Creating a single figure for each page
        fig, axes = plt.subplots(nrows=n_rows, ncols=n_columns, figsize=(11.7, 8.3))  # A4 size
        for i in range(total_plots):
            idx = page * total_plots + i
            if idx < len(columns):
                row = i // n_columns
                col = i % n_columns
                column = columns[idx]
                if plot_type == 'histogram':
                    sns.histplot(data=df, x=column, kde=True, ax=axes[row, col], bins=20)
                elif plot_type == 'boxplot':
                    sns.boxplot(data=df, x=column, ax=axes[row, col])

                # Set title for each subplot
                axes[row, col].set_title(f'{plot_type.capitalize()} of {column}')
            else:
                # Break if there are no more columns to plot
                break

        plt.tight_layout()
        plt.show()

print(len(df1))

# Plotting histograms
plot_subplots('histogram', df1)

# Plotting boxplots
plot_subplots('boxplot', df1)
```

```python
# Make a correlation matrix
corr_matrix = df1.corr()

# Make a heatmap of the correlation matrix
sns.heatmap(corr_matrix, annot=False)
plt.title('Correlation Matrix')
plt.show()
```
